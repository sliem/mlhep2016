```python
%matplotlib inline
import numpy
import pandas
import matplotlib.pyplot as plt
from matplotlib.pyplot import cm

import os, sys
sys.path.append('../code')
```

# Data Generation

```python
X = numpy.arange(0.1, 1, 0.2)
Y = 1. - 1. * X
```

# Radon Transform

```python
def RadonTransform(x, y):
    
    # s = r Cos(\phi - \alpha)
    r = numpy.sqrt(x**2 + y**2)
    phi = numpy.arccos(x / r)
    alpha = numpy.arange(-numpy.pi/2., numpy.pi/2., 0.001)
    s = r * numpy.cos(phi - alpha)
    
    return alpha, s
```

```python
plt.figure(figsize=(10, 7))
color=cm.rainbow(numpy.linspace(0,1,len(X)))

plt.plot(X, Y, color='0', linewidth=2, alpha=0.3)
for x,y,c in zip(X, Y, color):
    plt.scatter(x, y, color=c, linewidths=7)
plt.xlim(0., 1.)
plt.ylim(0., 1)
plt.xlabel('X', size=15)
plt.ylabel('Y', size=15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
all_alphas = []
all_s = []
for x,y,c in zip(X, Y, color):

    alpha, s = RadonTransform(x,y)
    all_alphas += list(alpha)
    all_s += list(s)

    plt.plot(alpha, s, color=c, linewidth=4)
    
plt.xlim(-numpy.pi/2., numpy.pi/2.)
plt.xlabel('alpha', size=15)
plt.ylabel('s', size=15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```

```python
plt.figure(figsize=(10, 7))
plt.hist2d(all_alphas, all_s, bins=50, cmin=0, normed=True)
plt.colorbar()
plt.xlim(-numpy.pi/2., numpy.pi/2.)
plt.xlabel('alpha', size=15)
plt.ylabel('s', size=15)
plt.xticks(size=15)
plt.yticks(size=15)
plt.show()
```
